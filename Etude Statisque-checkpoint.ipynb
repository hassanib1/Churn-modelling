```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import ttest_ind, chi2_contingency

# Charger les données
data = pd.read_csv('Documents//Churn_Modelling.csv')

# Analyse bivariée entre l'âge et la possession de carte de crédit
plt.figure(figsize=(10, 6))
sns.boxplot(x='HasCrCard', y='Age', data=data, palette="Set2")
plt.title("Distribution de l'âge en fonction de la possession d'une carte de crédit")
plt.xlabel("Possession de carte de crédit (1 = Oui, 0 = Non)")
plt.ylabel("Âge")
plt.show()

# Test de différence de moyenne entre les groupes (t-test)
age_with_card = data[data['HasCrCard'] == 1]['Age']
age_without_card = data[data['HasCrCard'] == 0]['Age']
t_stat, p_value = ttest_ind(age_with_card, age_without_card)
print(f"T-test entre l'âge des groupes avec et sans carte de crédit : p-value = {p_value:.4f}")
if p_value < 0.05:
    print("Il y a une différence significative d'âge entre les groupes avec et sans carte de crédit.")
else:
    print("Aucune différence significative d'âge entre les groupes avec et sans carte de crédit.")

# Analyse bivariée entre le churn (Exited) et le genre
plt.figure(figsize=(8, 5))
sns.countplot(x='Gender', hue='Exited', data=data, palette="Set1")
plt.title("Répartition du churn par genre")
plt.xlabel("Genre")
plt.ylabel("Nombre de clients")
plt.legend(title='Exited', labels=['Non (0)', 'Oui (1)'])
plt.show()

# Test du khi-deux pour le churn et le genre
contingency_table = pd.crosstab(data['Gender'], data['Exited'])
chi2, p, _, _ = chi2_contingency(contingency_table)
print(f"Test du khi-deux pour le churn et le genre : p-value = {p:.4f}")
if p < 0.05:
    print("Il y a une association significative entre le genre et le churn.")
else:
    print("Aucune association significative entre le genre et le churn.")

```

```python
import pandas as pd

# Charger vos données
data = pd.read_csv("Documents//Churn_Modelling.csv")  # Remplacez par le chemin de votre fichier

# Variables à analyser
variables = ['Age', 'CreditScore', 'Balance', 'EstimatedSalary']
target = 'Exited'

# Calcul des statistiques descriptives pour chaque variable
univariate_analysis = {}

for var in variables:
    grouped_stats = data.groupby(target)[var].agg([
        'mean',           # Moyenne
        'median',         # Médiane
        'min',            # Minimum
        'max',            # Maximum
        'std',            # Écart-type
        lambda x: x.quantile(0.25),  # 1er quartile
        lambda x: x.quantile(0.5),   # 2e quartile (médiane)
        lambda x: x.quantile(0.75),  # 3e quartile
        lambda x: x.quantile(1.0)    # 4e quartile (maximum)
    ])
    grouped_stats.columns = ['mean', 'median', 'min', 'max', 'std', '1st_quartile', '2nd_quartile', '3rd_quartile', '4th_quartile']
    univariate_analysis[var] = grouped_stats

# Afficher les résultats
for var, stats in univariate_analysis.items():
    print(f"Analysis for {var}:")
    print(stats)
    print("\n")
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Charger les données
data = pd.read_csv("Documents//Churn_Modelling.csv")  # Remplacez par le chemin de votre fichier

# Variables quantitatives continues
quantitative_vars = ['Age', 'CreditScore', 'Balance', 'EstimatedSalary']
target = 'Exited'

# Boxplots pour toutes les variables continues
plt.figure(figsize=(12, 8))
for i, var in enumerate(quantitative_vars, 1):
    plt.subplot(2, 2, i)  # Créer une grille de 2x2
    sns.boxplot(data=data, x=target, y=var, palette="Set2")
    plt.title(f'Boxplot of {var} by {target}')
plt.tight_layout()
plt.show()

# Densité pour une seule variable (exemple : 'Age')
plt.figure(figsize=(8, 6))
for value, color in zip(data[target].unique(), ['blue', 'orange']):
    subset = data[data[target] == value]
    sns.kdeplot(subset['Age'], fill=True, color=color, label=f'Exited={value}', alpha=0.5)
plt.title('Density Plot for Age by Exited')
plt.xlabel('Age')
plt.ylabel('Density')
plt.legend()
plt.show()
```
